```python
import os
from openai import OpenAI

# 1. Collect user input
print("Personalized Dinner Recipe Generator\n")
allergy_list       = input("Enter your allergies (e.g. peanuts, shellfish) or 'none': ").strip()
food_preferences   = input("Enter dietary preferences (e.g. vegetarian, keto, gluten-free) or 'none': ").strip()

if not allergy_list:
    allergy_list = "none"
if not food_preferences:
    food_preferences = "none"
```

```python
# ── 2. Build the prompt (f-string template) ────────────────────────────────────
prompt = f"""
You are a professional chef and nutritionist. Your task is to create a safe, delicious,
and personalized dinner recipe for a user based on their dietary needs.

USER PROFILE
- Allergies: {allergy_list}
- Dietary Preferences: {food_preferences}

INSTRUCTIONS
1. You MUST NOT include any ingredients the user is allergic to.
2. The recipe MUST comply with the user's dietary preferences.
3. The recipe should be practical for a home cook (ready in under 60 minutes).
4. Structure your response EXACTLY as shown in the example below.


EXAMPLE OUTPUT (for a vegan, nut-free user)
Recipe Name: Smoky Black Bean Tacos
Ingredients:
- 1 can black beans, drained
- 1 tsp smoked paprika
- 1/2 tsp cumin
- 4 small corn tortillas
- 1 avocado, sliced
- 1/2 cup salsa
- Fresh cilantro to taste

Instructions:
Step 1: Heat a pan over medium heat and add the drained black beans.
Step 2: Season with smoked paprika and cumin; stir and cook for 5 minutes.
Step 3: Warm the tortillas in a dry pan for 30 seconds each side.
Step 4: Assemble tacos with beans, avocado slices, salsa, and cilantro.


CHAIN-OF-THOUGHT REASONING — follow these steps internally before writing the recipe
Step A: Identify all ingredients that must be excluded due to the user's allergies: {allergy_list}.
Step B: Identify all dietary constraints from the user's preferences: {food_preferences}.
Step C: Think of a dinner dish that naturally fits these constraints without requiring major substitutions.
Step D: List the ingredients, confirming each one is safe and compliant.
Step E: Write clear, numbered cooking instructions a home cook can easily follow.
Step F: Give the recipe a creative, appetizing name.

Now generate the personalized recipe for this user.
""".strip()

#  3. Call the OpenAI API
api_key = ""
client  = OpenAI(api_key=api_key)

print("\nGenerating your recipe, please wait")

response = client.chat.completions.create(model    = "gpt-4o-mini", messages = [{"role": "user", "content": prompt}], temperature = 0.7,)

recipe_response = response.choices[0].message.content
```

```python
#  4. Display the result
print(recipe_response)

# 5. Save prompt + response to a .txt file
output_filename = "recipe_output.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write("GENERATED PROMPT")
    f.write(prompt)
    f.write("LLM RESPONSE")
    f.write(recipe_response)

print(f"Prompt and response saved to '{output_filename}'")
```
